```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dataset with 10 rows
data = {
    'ID': range(1, 11),
    'Group': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B', 'C', 'A'],
    'Value': [10, 12, 15, 14, 20, 22, 11, 16, 21, 13]
}
df = pd.DataFrame(data)
print("Population (Small Dataset):")
print(df)
```

```python
# Simple Random Sampling (Sample size = 4)
random_sample = df.sample(n=4, random_state=42)
print("\nSimple Random Sample:")
print(random_sample)
```

```python
# Stratified Sampling (Sample 1 from each Group)
stratified_sample = df.groupby('Group', group_keys=False).apply(lambda x: x.sample(1, random_state=42))
print("\nStratified Sample:")
print(stratified_sample)
```

```python
# Central Limit Theorem (CLT) Demonstration
# Since 10 rows are too few for CLT, we simulate a larger population
population = np.random.exponential(scale=2, size=10000)

sample_means = []
sample_size = 30
num_samples = 1000

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(10, 5))
sns.histplot(sample_means, kde=True)
plt.title('Sampling Distribution of the Mean (CLT Demo)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
print("The distribution of sample means approximates a normal distribution.")
```
