```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create a dataset with 10 rows
data = {
    'Income': [40, 45, 50, 55, 60, 65, 70, 75, 80, 85], # in 1000s
    'PurchaseAmount': [200, 220, 250, 270, 300, 320, 350, 370, 400, 420]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Visualize
plt.scatter(df['Income'], df['PurchaseAmount'])
plt.xlabel('Income')
plt.ylabel('Purchase Amount')
plt.title('Income vs Purchase Amount')
plt.show()
```

```python
# Model Building
X = df[['Income']]
y = df['PurchaseAmount']

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

# Evaluation
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
print(f"R2 Score: {r2_score(y, y_pred)}")
print(f"RMSE: {np.sqrt(mean_squared_error(y, y_pred))}")
```

```python
# Plot Regression Line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title('Regression Line')
plt.show()
```
