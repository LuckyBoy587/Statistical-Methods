```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
height_row = df["Height_cm"]
height_samples = []
for i in range(500):
    height_samples.append(height_row.sample(50).mean())
```

```python
sns.histplot(height_samples, bins=15, kde=True, color="brown")
plt.title("Sampling Distribution of the Mean (CLT in Action)")
plt.ylabel("Count")
plt.xlabel("Sample Mean Height (Cm) (n=50)")
plt.show()
```
