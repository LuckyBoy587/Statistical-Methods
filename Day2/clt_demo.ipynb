```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
```

```python
df = pd.read_excel(r'https://github.com/LuckyBoy587/Statistical-Methods/raw/refs/heads/main/data/ML374_S2_Height_Weight_Data_Concept.xlsx')
```

```python
heights = df['Height_cm'].dropna().values
```

```python
sample_size = 50
num_samples = 1000
sample_means = [np.mean(np.random.choice(heights, size=sample_size, replace=True)) for _ in range(num_samples)]
```

```python
sns.set_style('white')
fig, ax = plt.subplots(figsize=(6,5))
sns.histplot(sample_means, bins=14, kde=True, color='lightcoral', edgecolor='black', ax=ax)
ax.set_title('Sampling Distribution of the Mean (CLT in Action)', fontsize=16)
ax.set_xlabel('Sample Mean Height (Cm) (n=50)', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.2)
plt.tight_layout()
plt.show()
```
