```python
store_sales_url = r"https://drive.google.com/uc?export=download&id=1wzCEsY8QU-L0kPwbStxz-J1hvL24MBJD"
store_sales = pd.read_csv(store_sales_url)
store_sales.head()
```

```python
sample_sizes = {
    "Electronics": 2,
    "Grocery": 3,
    "Clothing": 2,
}
```

```python
sample_df = store_sales.groupby('Category', group_keys=False).apply(
    lambda grp: grp.sample(sample_sizes[grp.name], replace=False).assign(Category=grp.name),
    include_groups=False
).reset_index()
sample_df
```

```python
sample_df.describe()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population = np.random.exponential(scale=120, size=10000)
```

```python
plt.hist(population, bins=40, edgecolor='black')
plt.title('Population Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sample_means = []

for _ in range(1000):
    sample = np.random.choice(population, size=40)  
    sample_means.append(np.mean(sample))
    
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title('Sampling Distribution of the Sample Mean')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
