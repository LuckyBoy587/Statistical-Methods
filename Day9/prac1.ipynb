# Data Exploration and Visualization

In this notebook, we will explore the diabetes dataset to understand its structure and distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"https://github.com/LuckyBoy587/Statistical-Methods/raw/refs/heads/main/data/ML374_S9_Practice_Diabetes_Data.csv"
df = pd.read_csv(file_path)

# Display the first few rows
print("First 5 rows of the dataset:")
display(df.head())
```

```python
# Dataset info
print("Dataset Information:")
print(df.info())

# Summary statistics
print("Summary Statistics:")
display(df.describe())
```

```python
# Define a helper function to plot histograms
def plot_histogram(col):
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plot_histogram('Glucose')
```

```python
plot_histogram('BloodPressure')
```

```python
plot_histogram('SkinThickness')
```

```python
plot_histogram('Insulin')
```

```python
plot_histogram('BMI')
```

```python
plot_histogram('DiabetesPedigreeFunction')
```

```python
plot_histogram('Age')
```
