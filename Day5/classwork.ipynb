```python
import pandas as pd
import numpy as np

data = {
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
}
```

```python
df = pd.DataFrame(data)
print("Original DataFrame:")
df
```

```python
cleaned_df = df.dropna().reset_index(drop=True)
cleaned_df
```

```python
filled_df = df.fillna(df.mean(numeric_only=True).round(2))
filled_df
```
