```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=2500, centers=5)
```

```python
plt.scatter(X[:, 0], X[:, 1], c='orange', s=10)
```

```python
kmean = KMeans(n_clusters=5)
kmean.fit(X)
```

```python
centroids = kmean.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c='orange', s=10)
plt.scatter(centroids[:, 0], centroids[:, 1], c='blue', s=100, alpha=0.75)
plt.show()
```

```python
point = kmean.predict([[3.0, 1.5]])
centroids = kmean.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c='orange', s=10)
plt.scatter(centroids[:, 0], centroids[:, 1], c='blue', s=100, alpha=0.75)
plt.scatter(3.0, 1.5, c='red', s=200, marker='*')
plt.show()
print("The point [3.0, 1.5] belongs to cluster:", point[0])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import make_blobs

# Sample data
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Trying K = 2 to 10
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)

    score = silhouette_score(X, labels)
    print(f"K = {k}, Silhouette Score = {score:.4f}")
```
